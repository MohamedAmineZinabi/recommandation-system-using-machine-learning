```python
ratings = pd.read_csv('BX-Book-Ratings.csv', sep=";", on_bad_lines='skip', encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
x = ratings['User-ID'].value_counts() > 200
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['User-ID'].isin(y)]
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books, on="ISBN")
```

```python
ratings_with_books
```

```python
num_ratings = ratings_with_books.groupby('Book-Title')['Book-Rating'].count().reset_index()
```

```python
num_ratings.head()
```

```python
num_ratings.rename(columns={"Book-Rating":"num_of_ratings"}, inplace=True)
```

```python
num_ratings.head()
```

```python
final_rating = ratings_with_books.merge(num_ratings, on="Book-Title")
```

```python
final_rating.head()
```

```python
final_rating = final_rating[final_rating['num_of_ratings']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['User-ID', 'Book-Title'], inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='User-ID', index='Book-Title', values='Book-Rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
books_name = book_pivot.index
```

```python
import pickle
pickle.dump(model, open('artifacts/model.pkl', 'wb'))
pickle.dump(books_name, open('artifacts/books_name.pkl', 'wb'))
pickle.dump(final_rating, open('artifacts/final_rating.pkl', 'wb'))
pickle.dump(book_pivot, open('artifacts/book_pivot.pkl', 'wb'))
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)
    
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
book_name = 'Harry Potter and the Chamber of Secrets (Book 2)'
recommend_book(book_name)
```
